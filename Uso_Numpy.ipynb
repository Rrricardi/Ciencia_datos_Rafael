---

## 📚 Conceptos Clave de este Ejercicio

### 1. Operaciones Elemento a Elemento (Broadcasting)
Cuando usas operadores aritméticos estándar (`+`, `-`, `*`) entre dos matrices del mismo tamaño en NumPy, la operación se aplica **posición por posición**.
*   **Ejemplo:** Si `A[0,0] = 2` y `B[0,0] = 3`, entonces `(A + B)[0,0] = 5`.
*   **Uso:** Normalización, ajustes de brillo en imágenes, corrección de sesgos.

### 2. Producto Matricial (Matrix Multiplication)
Esta es la operación clásica del álgebra lineal (filas por columnas). En NumPy, **no** se usa el asterisco `*`.
*   **Sintaxis:** `A @ B` o `np.dot(A, B)`.
*   **Regla de Oro:** Para multiplicar $A_{n \times m}$ por $B$, las columnas de $A$ deben coincidir con las filas de $B$.
*   **Uso:** Capas de redes neuronales, transformación de coordenadas, sistemas de ecuaciones.

### 3. Encapsulamiento en Funciones
Al crear la función `operaciones_matrices`, estamos creando una herramienta reusable. Esto nos permite probar diferentes matrices sin reescribir la lógica matemática cada vez.

---

## ⚠️ Consideraciones Importantes

| Operación | Símbolo en NumPy | Requisito de Dimensiones |
| :--- | :---: | :--- |
| **Suma / Resta** | `+` / `-` | Las matrices deben ser **idénticas** en forma. |
| **Mult. Elemento** | `*` | Las matrices deben ser **idénticas** en forma (o compatibles por broadcasting). |
| **Producto Matricial** | `@` / `np.dot` | Columnas de $A$ == Filas de $B$. |

> 💡 **Consejo de Experto:** 
> Un error común es intentar hacer un producto matricial con `*`. ¡Cuidado! Eso te dará una multiplicación elemento a elemento. Usa siempre `@` para álgebra lineal real.

## 🚀 Tu Misión

Implementarás la función `operaciones_matrices(A, B)` que devolverá una tupla con 4 resultados:
1.  **Suma** (`A + B`)

Además, deberás manejar las dimensiones correctamente para evitar errores de `ValueError`. ¡Vamos a poner a prueba esas matrices!

```python
import numpy as np

n, m = 5, 5
A = np.random.rand(n, m)
B = np.random.rand(n, m)

def operaciones_matrices(A, B):
    producto = A @ B
    if A.shape == B.shape:
        suma = A + B
        resta = A - B
        multiplicacion = A * B
    return suma, resta, multiplicacion, producto

suma, resta, multiplicacion, producto = operaciones_matrices(A, B)

# ✅ Part 2: the missing matrix/vector tasks
matriz_100x100 = np.random.rand(100, 100)          # 1. 100×100 matrix
vector_aplanado = matriz_100x100.flatten()          # 2. flattened → 10,000 elements
cada_cien = vector_aplanado[::100]                  # 3. every 100th element → 100 elements
m999 = np.full((100, 100), 999)                     # 4. all values exactly 999

print("=" * 50)
print("SUMA (A + B):")
print(suma)
print("\n" + "=" * 50)
print("RESTA (A - B):")
print(resta)
print("\n" + "=" * 50)
print("MULTIPLICACIÓN ELEMENTO A ELEMENTO (A * B):")
print(multiplicacion)
print("\n" + "=" * 50)
print("PRODUCTO MATRICIAL (A @ B):")
print(producto)
print("=" * 50)
```

# 🖼️ Módulo 3: Manipulación de Imágenes Digitales con Python

¡Felicidades por llegar hasta aquí! Ahora que dominas la limpieza de datos y las operaciones matriciales, es momento de aplicar todo ese conocimiento a algo visual y tangible: **las imágenes digitales**.

¿Sabías que una imagen no es más que una **matriz de números**? Este ejercicio, `manipulación_imágenes`, te revelará el secreto detrás de cada píxel que ves en tu pantalla y te enseñará a procesar imágenes programáticamente.

## 🎯 Objetivos de Aprendizaje

1.  **Carga de Imágenes:** Usar **Pillow (PIL)** para abrir y preparar archivos de imagen.
2.  **Conversión a Array:** Transformar una imagen en una matriz de **NumPy** para manipulación numérica.
3.  **Entendiendo RGBA:** Comprender cómo los 4 canales (Rojo, Verde, Azul, Alpha) componen una imagen.
4.  **Análisis de Transparencia:** Identificar y contar píxeles opacos usando el canal Alpha.
5.  **Visualización:** Graficar cada canal individualmente con **Matplotlib**.

---

## 📚 Conceptos Clave de este Ejercicio

### 1. ¿Qué es una Imagen para Python?
Para nosotros, una imagen es solo una **matriz tridimensional** de números enteros:
*   **Alto (ny):** Número de filas de píxeles.
*   **Ancho (nx):** Número de columnas de píxeles.
*   **Canales:** Profundidad de la matriz (3 para RGB, 4 para RGBA).


```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

def analizar_imagen_y_graficar_canales(ruta_imagen):
    # ❌ Remove this line — it overrides the parameter!
    # ruta_imagen = "data/logo.png"
    
    imagen = Image.open(ruta_imagen).convert("RGBA")
    imagen_array = np.array(imagen)
    ny, nx = imagen_array.shape[:2]
    
    canal_alpha = imagen_array[:, :, 3]
    pixeles_no_transparentes = int(np.sum(canal_alpha == 255))
    
    fig, ejes = plt.subplots(1, 4, figsize=(18, 5))
    fig.suptitle("Canales de la Imagen", fontsize=16, fontweight="bold")
    
    nombres_canales = ["Rojo (R)", "Verde (G)", "Azul (B)", "Alpha (A)"]
    colormaps       = ["Reds",     "Greens",    "Blues",    "gray"]
    
    for i, (ax, nombre, cmap) in enumerate(zip(ejes, nombres_canales, colormaps)):
        canal = imagen_array[:, :, i]
        im = ax.imshow(canal, cmap=cmap, vmin=0, vmax=255)
        ax.set_title(nombre, fontsize=13)
        ax.axis("off")
        fig.colorbar(im, ax=ax, fraction=0.046, pad=0.04)
    
    plt.tight_layout()
    plt.savefig("canales_imagen.png", dpi=150, bbox_inches="tight")
    plt.show()
    
    print("Gráfica guardada como 'canales_imagen.png'")
    print(f"\n{'='*45}")
    print(f"  Ancho  (nx)                  : {nx} px")
    print(f"  Alto   (ny)                  : {ny} px")
    print(f"  Píxeles totales              : {nx * ny}")
    print(f"  Píxeles no transparentes     : {pixeles_no_transparentes}")
    print(f"  Porcentaje opaco             : {pixeles_no_transparentes / (nx * ny) * 100:.1f}%")
    print(f"{'='*45}\n")
    
    return nx, ny, pixeles_no_transparentes

if __name__ == "__main__":
    RUTA = "data/logo.png"
    nx, ny, pixeles_no_transparentes = analizar_imagen_y_graficar_canales(RUTA)
```

## 📚 Conceptos Clave de este Ejercicio

### 1. De Vectores a Mallas: `np.meshgrid`
Para graficar una superficie, no basta con tener dos vectores `x` e `y`. Necesitamos **todas las combinaciones posibles** de pares $(x_i, y_j)$.


```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def graficar_funcion():
    # Vectores base
    x = np.linspace(0, 10, 100)      # 100 elementos
    y = np.linspace(-1, 1, 200)      # 200 elementos
    
    # Malla con forma (200, 100): filas=y, columnas=x
    X, Y = np.meshgrid(x, y)         # X.shape = Y.shape = (200, 100)
    
    # Calcular Z
    Z = X**2 + 3/Y + 3
    
    return X, Y, Z

X, Y, Z = graficar_funcion()

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection="3d")
ax.plot_surface(X, Y, Z, rstride=4, cstride=4,
                linewidth=0, antialiased=False, cmap='viridis')
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.set_title("Superficie: Z = X² + 3/Y + 3")
plt.show()
```
