```python
# your code here
import pandas as pd
import matplotlib.pyplot as plt
cuerna = pd.read_csv('data/Cuernavaca_Enero_comas.csv', index_col=0, parse_dates=True)

# your code here
fecha_inicio=pd.to_datetime('2012-01-01')
fecha_fin=fecha_inicio+ pd.Timedelta('3.5d')
fig, ax2= plt.subplots(3,figsize=(10,6), sharex=True)

ax2[0].plot(cuerna["To"], label="To")
ax2[0].set_ylabel('Temperatura (°C)')
ax2[0].legend()
ax2[0].set_xlim(fecha_inicio, fecha_fin)

ax2[1].plot(cuerna['RH'], color='green', label='HR')
ax2[1].set_ylabel('Humedad Relativa (%)')
ax2[1].legend()
ax2[1].set_xlim(fecha_inicio, fecha_fin)

ax2[2].plot(cuerna['Ig'], color='red', label='Ig')
ax2[2].plot(cuerna['Ib'], color='blue', label='Ib')
ax2[2].plot(cuerna['Id'], color='purple', label='Id')
ax2[2].set_ylabel('Radiación solar [W/m2]')
ax2[2].set_xlabel('Tiempo')
ax2[2].legend()
ax2[2].set_xlim(fecha_inicio, fecha_fin)

plt.tight_layout()
plt.show()
```

Sigue cada paso detenidamente para asegurar la correcta configuración y visualización de los gráficos. Cada detalle, desde las etiquetas hasta los ajustes de los ejes, contribuye significativamente a la calidad y precisión de la visualización final.

### La imagen resultante debe ser así:


```python
# Suponiendo que 'cuerna' es tu DataFrame y ya está cargado correctamente
# grafica de acuerdo a las instrucciones.
# your code here
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
from matplotlib.dates import DateFormatter
cuerna = pd.read_csv('data/Cuernavaca_Enero_comas.csv', index_col=0, parse_dates=True)
fecha_inicio=pd.to_datetime('2012-01-01')
fecha_fin=fecha_inicio+ pd.Timedelta('3.5d')


cuerna_filtrado = cuerna[fecha_inicio:fecha_fin]

fig=plt.figure(figsize=(10,5))
gs=GridSpec(2,4, figure=fig)
ax1= fig.add_subplot(gs[0,0:2])
ax2= fig.add_subplot(gs[1,0:2], sharex=ax1)
ax3 = fig.add_subplot(gs[:, 2:4])
#Primer Subgráfico - Temperatura
ax1.plot(cuerna_filtrado['To'], label='To')
ax1.set_ylabel('Temperatura (°C)')
ax1.legend()
ax1.set_xlim(fecha_inicio, fecha_fin)

#Segundo Subgráfico - Humedad Relativa1.tick_params(labelbottom=False)  

ax2.plot(cuerna['RH'], color='green', label='HR')
ax2.set_ylabel('Humedad Relativa (%)')
ax2.set_xlabel('Tiempo')
ax2.legend()
ax2.set_xlim(fecha_inicio, fecha_fin)
# Aplicar formato de fecha %m-%d al eje x para mejor legibilidad
ax2.xaxis.set_major_formatter(DateFormatter('%m-%d'))


# Tercer Subgráfico - Dispersión Temperatura vs Humedad

ax3.scatter(cuerna['To'], cuerna['RH'], alpha=0.5, label='To' )
ax3.set_xlabel('Temperatura (°C)')
ax3.set_ylabel('Humedad Relativa (%)')

plt.tight_layout()
plt.show()
```
